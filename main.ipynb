```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Attrition data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
print(df.duplicated().value_counts())
df.drop_duplicates(inplace=True)
print(len(df))
```

```python
df['JobSatisfaction'].mode()
```

```python
df['JobSatisfaction'].fillna(4,inplace=True)
```

```python
df['WorkLifeBalance'].mode()
```

```python
df['WorkLifeBalance'].fillna(3,inplace=True)
```

```python
df['NumCompaniesWorked'].mode()
```

```python
df['NumCompaniesWorked'].fillna(1,inplace=True)
```

```python
df['TotalWorkingYears'].mode()
```

```python
df['TotalWorkingYears'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,5))
plt.rc('font',size=14)
sns.countplot(x='Attrition', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.rc('font',size=14)
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title('Attrition by Gender')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.show()
```

```python
df.groupby(["Gender", "Attrition"]).agg({'Age': 'mean'})
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='EducationField', hue='Attrition', data=df)
plt.title('Attrition by EducationField')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Age'],hist=False)
plt.show()
```

```python
ordinal_features = ['Education','EnvironmentSatisfaction','JobInvolvement','JobSatisfaction',
                    'PerformanceRating','WorkLifeBalance']
df[ordinal_features].head()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Education', hue='Attrition', data=df)
plt.title('Attrition by Education ')
plt.show()
```
